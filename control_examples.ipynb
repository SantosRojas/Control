# Ejemplos de uso del modulo control

```python
#importamos la libreria, la cual deberia estar en la misma carpeta de trabajo o en todo caso agregarla al path
import control as cn
%matplotlib notebook
```

```python
#Declaramos las funciones de transferencia
sys_a=cn.tf(1,[0.5,1],0.5)
sys_b=cn.tf(2,[0.2,2])
```

```python
#Serie de dos tf
sys_serie=cn.serie(sys_a,sys_b)
sys_serie
```

```python
#Feedback
sys_f=cn.tf(2,1)
sys_feedback=cn.feedback(sys_a,sys_f)
sys_feedback
```

```python
#Ploteo simple
cn.plot_step([sys_a,{'label':'Mi funcion 1'}],[sys_b,{'label':'Mi funcion 2'}],sys_feedback)
```

```python
#Ploteo modificando propiedades de cada sistema
cn.plot_step([sys_a,{'c':'m--','lw':2,'label':'Respuesta del orno'}],[sys_b,{'c':'c:','lw':3,'label':'Mi funcion 2'}],T=[0,3.5,100])
```

```python
#Ploteo con parametros
sys1=cn.tf(2,[1,0.8,2],0.5)#tf de segundo orden con retardo de 0.5 segundos
cn.plot_step(sys_a,sys1,params='all',legend_params=True)
```

```python
#Mostrando solo los parametros
cn.plot_step(sys_a,sys_b,sys_feedback,sys_serie,only_params=True,params='all')
```

```python
#ploteo de varios sistemas definiendo el vector de tiempo T
cn.plot_step(sys_a,sys_b,sys_feedback,sys_serie,
             T=[0,5,100],
             title='Mi grafica',
             xlabel='x(s)',
             ylabel='Y(T)')
```
